### Question 1

Load data in fraude/movimientos_tarjetas_black_caja_madrid.csv

```python
from pyspark import SparkContext, SQLContext
sc =SparkContext()
sqlContext = SQLContext(sc)

def answer_one():
    df = sqlContext.read.format('com.databricks.spark.csv') \
                    .options(header='true', inferschema='true', delimiter=';') \
                    .load('file:///home/jovyan/work/Presentacion/fraude/movimientos_tarjetas_black_caja_madrid.csv')
    return df

df=answer_one() 
df.printSchema()
```

### Question 2

Obtain the 10 executives with the highest amount of expenses

```python
import pandas as pd
def answer_four():
    df_mapped=df.select('nombre','importe').rdd.map(lambda x: (x[0],x[1])) 
    df_reduced=df_mapped.reduceByKey(lambda x,y:x+y)
    df_top=df_reduced.map(lambda x:(x[1],x[0])).sortByKey(ascending=False)
    return df_top

executives=answer_four()

executives=pd.DataFrame(executives.take(10), columns=["Gasto", "Ejecutivo"]) 
executives
```

### Question 3

Obtain the 10 persons who have expenses in businesses with the activity "CLUBS,SALAS FIESTA,PUBS,DISCOTECAS,BARES" and the total amount spent 

```python
def answer_five():
    df_expenses = df.filter("actividad LIKE '%CLUBS%'" 
                    or "actividad LIKE '%SALAS FIESTA%'" 
                    or "actividad LIKE '%PUBS%'" 
                    or "actividad LIKE '%SDISCOTECAS%'"
                    or "actividad LIKE '%BARES%'")
    df_expenses_mapped = df_expenses.rdd.map(lambda x: (x.nombre,x.importe))
    df_expenses_reduced = df_expenses_mapped.reduceByKey(lambda x,y:x+y)
    df_expenses_top=df_expenses_reduced.map(lambda x:(x[1], x[0])).sortByKey(ascending=False).take(10)
    
    return pd.DataFrame([ (x[1],x[0]) for x in df_expenses_top ] ,columns=['name','total_amount'])
    
persons=answer_five() 
persons
```
